## Configure plotter

```python
def create_test_inputs(x_min=[-3, -3], x_max=[3, 3], input_dim=4, num_test: int = 1600, factor: float = 1.2):
    sqrtN = int(np.sqrt(num_test))
    xx = np.linspace(x_min[0] * factor, x_max[0] * factor, sqrtN)
    yy = np.linspace(x_min[1] * factor, x_max[1] * factor, sqrtN)
    xx, yy = np.meshgrid(xx, yy)
    test_inputs = np.column_stack([xx.reshape(-1), yy.reshape(-1)])
    if input_dim > 2:
        zeros = np.zeros((num_test, input_dim - 2))
        test_inputs = np.concatenate([test_inputs, zeros], -1)
    return test_inputs, xx, yy

test_inputs, xx, yy = create_test_inputs()
plotter = ModeOptContourPlotter(mode_optimiser, test_inputs=test_inputs, static=False, explorative=True)
```

```python
def plot_trajectories_over_desired_mixing_prob(fig, ax, iteration):
    mixing_probs = mode_optimiser.dynamics.mosvgpe.gating_network.predict_mixing_probs(test_inputs)
    if iteration == 0:
        label="$\Pr(\\alpha=" + str(mode_optimiser.desired_mode + 1) + " \mid \mathbf{x}, \mathcal{D}_{0})$"
    else:
        label="$\Pr(\\alpha=" + str(mode_optimiser.desired_mode + 1) + " \mid \mathbf{x}, \mathcal{D}_{0:" + str(iteration) + "})$"
    plotter.mosvgpe_plotter.plot_contf(ax, mixing_probs[:, mode_optimiser.desired_mode], label=label)
#     plotter.plot_trajectories_given_fig(fig, plot_satisfaction_prob=True)
    for key in plotter.trajectories.keys():
        plotter.plot_trajectory_given_ax(ax, plotter.trajectories[key], key)
    return fig

def plot_trajectories_over_gating_function_variance(fig, ax, iteration):
    h_means, h_vars = mode_optimiser.dynamics.mosvgpe.gating_network.predict_h(test_inputs)
    if iteration == 0:
        label="$\mathbb{V}[h_{" + str(mode_optimiser.desired_mode + 1) + "}(\mathbf{x}) \mid \mathcal{D}_{0}]$"
    else:
        label="$\mathbb{V}[h_{" + str(mode_optimiser.desired_mode + 1) + "}(\mathbf{x}) \mid \mathcal{D}_{0:" + str(iteration) + "}]$"
    plotter.mosvgpe_plotter.plot_contf(ax, h_vars[:, mode_optimiser.desired_mode], label=label)
#     plotter.plot_trajectories_given_fig(fig, plot_satisfaction_prob=True)
#     trajectories = plotter.generate_trajectories()
    for key in plotter.trajectories.keys():
        plotter.plot_trajectory_given_ax(ax, plotter.trajectories[key], key)
    return fig

def plot_trajectories_given_ax(ax):
    plotter.generate_trajectories()
    for key in plotter.trajectories.keys():
        plotter.plot_trajectory_given_ax(ax, plotter.trajectories[key], key)

def plot_desired_mixing_prob(fig, ax):
    mixing_probs = mode_optimiser.dynamics.mosvgpe.gating_network.predict_mixing_probs(test_inputs)
    contf = ax.tricontourf(test_inputs[:, 0], 
                           test_inputs[:, 1], 
                           mixing_probs[:, mode_optimiser.desired_mode], 
                           levels=10,
                           cmap=plotter.mosvgpe_plotter.cmap)
    return contf

def plot_gating_function_variance(fig, ax):
    _, h_vars = mode_optimiser.dynamics.mosvgpe.gating_network.predict_h(test_inputs)
    contf = ax.tricontourf(test_inputs[:, 0], 
                           test_inputs[:, 1], 
                           h_vars[:, mode_optimiser.desired_mode], 
                           levels=10,
                           cmap=plotter.mosvgpe_plotter.cmap)
    return contf

def plot_data_over_gating_function_variance(fig, ax, iteration, epoch):
    h_means, h_vars = mode_optimiser.dynamics.mosvgpe.gating_network.predict_h(test_inputs)
    if iteration == 0:
        label="$\mathcal{D}_{0} \\text{ over } \mathbb{V}[h_{" + str(mode_optimiser.desired_mode + 1) + "}(\mathbf{x}) \mid \mathcal{D}_{0}]$"
    else:
        if epoch == 0:
            if iteration-1 == 0:
                label="$\mathcal{D}_{0:" + str(iteration) + "} \\text{ over } \mathbb{V}[h_{" + str(mode_optimiser.desired_mode + 1) + "}(\mathbf{x}) \mid \mathcal{D}_{0}]$"
            else:
                label="$\mathcal{D}_{0:" + str(iteration) + "} \\text{ over } \mathbb{V}[h_{" + str(mode_optimiser.desired_mode + 1) + "}(\mathbf{x}) \mid \mathcal{D}_{0:" + str(iteration-1) + "}]$"
        else:
            label="$\mathcal{D}_{0:" + str(iteration) + "} \\text{ over } \mathbb{V}[h_{" + str(mode_optimiser.desired_mode + 1) + "}(\mathbf{x}) \mid \mathcal{D}_{0:" + str(iteration) + "}]$"
    plotter.mosvgpe_plotter.plot_contf(ax, h_vars[:, mode_optimiser.desired_mode], label=label)
#     plotter.plot_data_given_fig(fig)
    return fig
```

## Plot ModeOpt 

```python
def plot_iteration(fig, axs, i, plot_contf_fn, label, row):
#     checkpoint.restore(os.path.join(ckpt_dir, "ckpt-"+str(i+2)))
    checkpoint.restore(os.path.join(ckpt_dir, "ckpt-"+str(i+1)))
    mode_optimiser.dataset = (full_dataset[0][:int(mode_optimiser.dataset_idx.numpy()), :], 
                              full_dataset[0][:int(mode_optimiser.dataset_idx.numpy()), :])
    checkpoint.restore(os.path.join(ckpt_dir, "ckpt-"+str(i+2)))

    for ax in axs.flat:
        contf = plot_contf_fn(fig, ax)
        plotter.plot_mode_satisfaction_probability_given_ax(ax)
    divider = make_axes_locatable(axs[-1])
    cax = divider.append_axes("right", size="5%", pad=0.05)
    cbar = plt.colorbar(contf, use_gridspec=True, cax=cax)
    if label == "desired_prob":
        if i == 0:
            cbar.set_label("$\Pr(\\alpha=" + str(mode_optimiser.desired_mode + 1) + " \mid \mathbf{x}, \mathcal{D}_{0})$")
        else:
            cbar.set_label("$\Pr(\\alpha=" + str(mode_optimiser.desired_mode + 1) + " \mid \mathbf{x}, \mathcal{D}_{0:" + str(i) + "})$")
    else:
        if i == 0:
            cbar.set_label("$\mathbb{V}[h_{" + str(mode_optimiser.desired_mode + 1) + "}(\mathbf{x}) \mid \mathbf{x}, \mathcal{D}_{0})$")
        else:
            cbar.set_label("$\mathbb{V}[h_{" + str(mode_optimiser.desired_mode + 1) + "}(\mathbf{x}) \mid \mathbf{x}, \mathcal{D}_{0:" + str(i) + "}]$")

    print(f'i {i}')
    print(mode_optimiser.dataset[0].shape)
    if i == 0:
        mode_optimiser.dataset = (full_dataset[0][:123, :], full_dataset[0][:123, :])
    elif row == 0:
        mode_optimiser.dataset = (full_dataset[0][:int(mode_optimiser.dataset_idx.numpy()), :], 
                                  full_dataset[0][:int(mode_optimiser.dataset_idx.numpy()), :])
    checkpoint.restore(os.path.join(ckpt_dir, "ckpt-"+str(i+2)))
#     checkpoint.restore(os.path.join(ckpt_dir, "ckpt-"+str(i+2)))
#     checkpoint.restore(os.path.join(ckpt_dir, "ckpt-"+str(i+3)))
    plotter.plot_data_given_ax(axs[0])
    mode_optimiser.dataset = (full_dataset[0][:int(mode_optimiser.dataset_idx.numpy()), :], 
                              full_dataset[0][:int(mode_optimiser.dataset_idx.numpy()), :])
    print(mode_optimiser.dataset[0].shape)
    plotter.plot_data_given_ax(axs[-1])
    plot_trajectories_given_ax(axs[1])
    axs[0].set_ylabel(f"$y \quad (i={i})$")


def plot_specific_iterations(plot_contf_fn, iterations, label, save_filename=None):
    fig = plt.figure(figsize=(plotter.mosvgpe_plotter.figsize[0] * 1.2, plotter.mosvgpe_plotter.figsize[1]*2.5))
    gs = fig.add_gridspec(4, 3, wspace=0.0, hspace=0.0)
    axs = gs.subplots(sharex="col", sharey="row")

    row = 0
    for i in iterations:
        plot_iteration(fig, axs[row, :], i, plot_contf_fn=plot_contf_fn, label=label, row=row)
        if label == "desired_prob":
            axs[0, 0].set_title("$\mathcal{D}_{0:i}$ over $\Pr(\\alpha=" + str(mode_optimiser.desired_mode + 1) + " \mid \mathbf{x}, \mathcal{D}_{0:i})$")
            axs[0, 2].set_title("$\mathcal{D}_{0:i+1}$ over $\Pr(\\alpha=" + str(mode_optimiser.desired_mode + 1) + " \mid \mathbf{x}, \mathcal{D}_{0:i})$")
        else:
            axs[0, 0].set_title("$\mathcal{D}_{0:i}$ over $\mathbb{V}[h_{" + str(mode_optimiser.desired_mode + 1) + "}(\mathbf{x}) \mid \mathbf{x}, \mathcal{D}_{0:i}]$")
            axs[0, 2].set_title("$\mathcal{D}_{0:i+1}$ over $\mathbb{V}[h_{" + str(mode_optimiser.desired_mode + 1) + "}(\mathbf{x}) \mid \mathbf{x}, \mathcal{D}_{0:i}]$")

        axs[0, 1].set_title("$\pi^i_{\\text{explore}}$ used to collect $\mathcal{D}_{i+1}$")
        row += 1
    
    for ax in axs[-1, :].flat:
        ax.set_xlabel("$x$")

    plotter.plot_env_no_obs_start_end_given_fig(fig)
    handles, labels = axs[-1, 1].get_legend_handles_labels()
    handles_obs, labels_obs = axs[-1, -1].get_legend_handles_labels()
    labels = labels + labels_obs
    handles = handles + handles_obs
    by_label = OrderedDict(zip(labels, handles))

    fig.tight_layout()
    fig.legend(
        by_label.values(),
        by_label.keys(),
        bbox_to_anchor=(0.5, -0.018),
        loc="lower center",
        bbox_transform=fig.transFigure,
        ncol=len(by_label),
    )
    if save_filename:
        plt.savefig(save_filename, transparent=True)
```
